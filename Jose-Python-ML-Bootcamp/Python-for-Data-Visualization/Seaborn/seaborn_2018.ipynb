# Seaborn
Seaborn is statistical plotting library. It has beautiful default styles. It is also designed to work very well with pandas dataframe object.
**Conda install seaborn**

### Distribution plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'],bins=30,)
```

```python
sns.distplot(tips['total_bill'],bins=100,kde=False)
```

```python
sns.jointplot(x=tips['total_bill'],y = tips['tip'],kind='scatter')
```

```python
sns.jointplot(x=tips['total_bill'],y = tips['tip'],kind='reg')
```

```python
sns.jointplot(x=tips['total_bill'],y = tips['tip'],kind='hex')
```

```python
sns.jointplot(x=tips['total_bill'],y = tips['tip'],kind='kde')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data =tips)
sns.swarmplot(x = 'day', y = 'total_bill', data =tips,color='yellow')
```

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips, kind='violin')
```

## Matrix Plot

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tc = tips.corr()
sns.heatmap(tc, annot= True, cmap ='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=3)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

## Regression plot

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers =['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col='day',hue='sex', aspect=0.6, size=8)
```

## Grids

```python

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(plt.scatter,'total_bill','tip')
```

## Style and color

```python
plt.figure(figsize=(12,3))
sns.set_style('ticks')
sns.countplot(x = 'sex', data = tips)
sns.despine()
```

```python
sns.set_context('notebook')
sns.set_style('ticks')
sns.countplot(x = 'sex', data = tips)
sns.despine()
```

```python
sns.lmplot('total_bill', y = 'tip', data = tips, hue='sex', palette='Wistia')
```

## Seaborn Exercises

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.jointplot(x = 'fare', y = 'age', data = titanic)
```

```python
sns.set_style('whitegrid')
sns.distplot(titanic['fare'],kde=0,color='r')
```

```python
sns.boxplot(x = 'class',y = 'age', data = titanic,palette='coolwarm')
```

```python
sns.swarmplot(x = 'class',y = 'age', data = titanic, palette = 'rainbow')
```

```python
sns.countplot(titanic['sex'])
```

```python
tc = titanic.corr()
sns.heatmap(tc,cmap='coolwarm')
```

```python
g = sns.FacetGrid(titanic,col='sex')
g.map(plt.hist,'age')
```
