```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,15,11)
y = x**2
x
```

```python
# Functional method
plt.plot(x,y,'r--')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Title')
```

```python
#Multi plot
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'b--')
```

```python
# Object orietened method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'g--')
axes.set_xlabel('$x$')
axes.set_ylabel('$y$')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.2,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'y--')
axes1.set_title('Larege plot')
axes2.set_title('Smaller plot')
```

### Matplotlib part2

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'b--')
axes[0].set_title('First plot')
axes[1].plot(x,y,'g--')
axes[1].set_title('Second plot')
plt.tight_layout()
```

### Figure size and dpi

```python
fig = plt.figure(figsize=(10,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'y-*')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(10,6))

axes[0].plot(x,y,'g--')
axes[0].set_xlabel('$X$')
axes[0].set_ylabel('$Y$')
axes[1].plot(y,x,'r-*')
axes[1].set_ylabel('$X$')
axes[1].set_xlabel('$Y$')
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=1200)
```

```python
fig = plt.figure(figsize=(10,4))

plt.plot(x,x**2, label='$x^2$')
plt.plot(x,x**3, label='$x^3$')
plt.legend()
```

### Matplotlib part3

```python
# Plot appearance
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00') # RGB Hex code
ax.plot(x,y**2, color='purple')
```

```python
# Plot appearance
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',linewidth=3,alpha=0.5) # RGB Hex code, alpha transperant
```

```python
# Plot appearance
fig = plt.figure(figsize=(10,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',lw=3, linestyle='steps')
```

```python
# Marker
fig = plt.figure(figsize=(10,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',lw=3, linestyle=':', marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```

```python
# plot range
fig =plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple', lw=2, ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
