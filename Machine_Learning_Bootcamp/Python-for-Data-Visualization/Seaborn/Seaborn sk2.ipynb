```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = pd.read_csv('My_Data.csv')
```

```python
tips.head()
```

```python
sns.distplot(tips['Profit'],kde=False,bins=5)
```

```python
sns.jointplot(x='R&D Spend',y='Profit',data=tips,kind='reg')
```

```python
sns.pairplot(tips,hue='State',palette='coolwarm')
```

```python
sns.rugplot(tips['Profit'])
```

```python
sns.kdeplot(tips['Profit'])
```

Categorical plot

```python
import numpy as np
```

```python
sns.barplot(x='State',y='Profit',data=tips,estimator=np.std)
```

```python
sns.countplot(x='State',data=tips)
```

```python
sns.boxplot(x='State',y='Profit',data=tips)
```

```python
sns.violinplot(x='State',y='Profit',data=tips)
```

```python
sns.stripplot(x='State',y='Profit',data=tips,jitter=True)
```

```python
sns.violinplot(x='State',y='Profit',data=tips)
sns.swarmplot(x='State',y='Profit',data=tips,color='black')
```

```python
sns.factorplot(x='State',y='Profit',data=tips,kind='bar')
```

Matrix plot

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
sns.clustermap(tc)
```

Regression plot

```python
sns.lmplot(x='Profit',y='R&D Spend',data=tips,hue='State',markers=['o','v','s'])
```

```python
sns.lmplot(x='Profit',y='R&D Spend',data=tips,col='State')
```

Grid

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
