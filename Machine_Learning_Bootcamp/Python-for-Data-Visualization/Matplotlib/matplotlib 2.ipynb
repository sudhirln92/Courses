Data visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y,'r-')
plt.xlabel('X lable')
plt.ylabel('Y lable')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented method
fig = plt.figure()

axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y lable')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.2,.5,.4,.3])

axes1.plot(x,y)
axes1.set_title('Larges plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

Matplotlib part 2

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')
plt.tight_layout()
```

Figure size and DPI

```python
fig = plt.figure(figsize=(3,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X^2')
ax.plot(x,x**3,label='X^3')

ax.legend(loc=0)
```

Plot appearance 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth= 0.5, linestyle='--',marker='o',markersize=5,
       markerfacecolor='red')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

plt.show()
```
